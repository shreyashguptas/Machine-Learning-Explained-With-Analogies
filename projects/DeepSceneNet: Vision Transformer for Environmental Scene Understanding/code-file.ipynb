# DeepSceneNet: Vision Transformer for Environmental Scene Understanding

## Project Overview
Build a deep learning model that can classify different types of scenes/environments (indoor, outdoor, urban, natural, etc.) from images, while also implementing visualization techniques to understand what features the model uses to make its decisions.

## Dataset
Use the Intel Image Classification dataset available on Kaggle:
https://www.kaggle.com/datasets/puneet6060/intel-image-classification

This dataset contains around 25,000 images (size 150x150) spread across 6 categories:
- Buildings
- Forest 
- Glacier
- Mountain
- Sea
- Street

## Project Goals
- Build and train a CNN model for scene classification
- Implement GradCAM visualization to understand model decisions
- Compare performance of different architectures
- Create an interpretable and reproducible solution

## Technical Requirements
- Use PyTorch for model implementation
- Implement proper training/validation/test splits
- Use data augmentation to improve generalization
- Implement learning rate scheduling
- Add dropout and regularization to prevent overfitting
- Create visualizations of training progress
- Generate GradCAM heatmaps for model interpretability

## Project Structure

### 1. Data Loading & Preprocessing
- Load and organize the dataset
- Implement data augmentation
- Create data loaders

### 2. Model Architecture
- Design a custom CNN architecture
- Implement transfer learning using pre-trained models (ResNet, VGG)
- Add proper regularization techniques

### 3. Training Pipeline
- Create training loop with validation
- Implement learning rate scheduling
- Add early stopping
- Track and plot metrics

### 4. Visualization & Interpretation
- Implement GradCAM visualization
- Create confusion matrices
- Generate example predictions with heatmaps

### 5. Model Evaluation
- Test set performance metrics
- Compare different architectures
- Analyze failure cases
